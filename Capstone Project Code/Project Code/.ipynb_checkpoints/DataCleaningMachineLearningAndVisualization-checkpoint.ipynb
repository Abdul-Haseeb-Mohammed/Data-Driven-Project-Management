```python
X = df_filtered.drop('Actual Cost', axis=1)
y = df_filtered['Actual Cost']

X_train, X_test, y_train, y_test, train_indices, test_indices = train_test_split(
    X, y, X.index, test_size=0.2, stratify=X['Portfolio WBS_encoded'], random_state=1
)
```

```python
dtc_regressor = DecisionTreeRegressor(random_state=1)
param_grid = {
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search_dt = GridSearchCV(estimator=dtc_regressor, param_grid=param_grid, cv=25, scoring='neg_mean_absolute_error', n_jobs=-1)
grid_search_dt.fit(X_train, y_train)

best_params = grid_search_dt.best_params_
best_score = -grid_search_dt.best_score_

print("Best Parameters:", best_params)
print("Best Mean Absolute Error:", best_score)
```

```python
y_pred_test_dt = grid_search_dt.predict(X_test)
y_pred_train_dt = grid_search_dt.predict(X_train)
```

```python
y_pred_combined = pd.Series(index=df2.index, dtype=float)

y_pred_combined.loc[train_indices] = y_pred_train_dt
y_pred_combined.loc[test_indices] = y_pred_test_dt

df2['predicted_DT'] = y_pred_combined

print(df2.head())
```

```python
rf_regressor = RandomForestRegressor(random_state=1)
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

grid_search_rf = GridSearchCV(estimator=rf_regressor, param_grid=param_grid, cv=25, scoring='neg_mean_absolute_error', n_jobs=-1)
grid_search_rf.fit(X_train, y_train)

best_params_rf = grid_search_rf.best_params_
best_score_rf = -grid_search_rf.best_score_

print("Best Parameters:", best_params_rf)
print("Best Mean Absolute Error:", best_score_rf)
```

```python
y_pred_test_rf = grid_search_rf.predict(X_test)
y_pred_train_rf = grid_search_rf.predict(X_train)
```

```python
y_pred_combined = pd.Series(index=df2.index, dtype=float)

y_pred_combined.loc[train_indices] = y_pred_train_rf
y_pred_combined.loc[test_indices] = y_pred_test_rf

df2['predicted_RF'] = y_pred_combined

print(df2.head())
```

```python
mae_dt_test = mean_absolute_error(y_test, y_pred_test_dt)
mae_dt_train = mean_absolute_error(y_train, y_pred_train_dt)

mse_dt_test = mean_squared_error(y_test, y_pred_test_dt)
mse_dt_train = mean_squared_error(y_train, y_pred_train_dt)

r2_dt_test = r2_score(y_test, y_pred_test_dt)
r2_dt_train = r2_score(y_train, y_pred_train_dt)

print("Decision Tree Regressor Metrics:")
print(f"Mean Absolute Error (Test): {mae_dt_test}")
print(f"Mean Absolute Error (Train): {mae_dt_train}")
print(f"Mean Squared Error (Test): {mse_dt_test}")
print(f"Mean Squared Error (Train): {mse_dt_train}")
print(f"R-squared (Test): {r2_dt_test}")
print(f"R-squared (Train): {r2_dt_train}")
```

```python
X = df2.drop('Actual Cost', axis=1)
y = df2['Actual Cost']
```

```python
categorical_columns = ['Portfolio WBS', 'UOM', 'Month']
numeric_columns = X.select_dtypes(include=['int64', 'float64']).columns
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_columns), 
        ('cat', OneHotEncoder(), categorical_columns)
    ])

X_preprocessed = preprocessor.fit_transform(X)

portfolio_wbs = X['Portfolio WBS']

X_train, X_test, y_train, y_test, train_indices, test_indices  = train_test_split(
    X_preprocessed, y, X.index, test_size=0.2, random_state=1, stratify=portfolio_wbs
)
```

```python
param_grid = {
    'fit_intercept': [True, False],
    'copy_X': [True, False],
    'n_jobs': [None, -1],
    'positive': [False]
}
linear_regressor = LinearRegression()

grid_search_lr = GridSearchCV(estimator=linear_regressor, param_grid=param_grid, cv=25, scoring='neg_mean_absolute_error', n_jobs=-1)
grid_search_lr.fit(X_train, y_train)

best_params = grid_search_lr.best_params_
best_score = -grid_search_lr.best_score_

print("Best Parameters:", best_params)
print("Best Mean Absolute Error:", best_score)
```

```python
y_pred_train = grid_search_lr.predict(X_train)
y_pred_test = grid_search_lr.predict(X_test)

mae_test = mean_absolute_error(y_test, y_pred_test)
mae_train = mean_absolute_error(y_train, y_pred_train)

mse_test = mean_squared_error(y_test, y_pred_test)
mse_train = mean_squared_error(y_train, y_pred_train)

r2_test = r2_score(y_test, y_pred_test)
r2_train = r2_score(y_train, y_pred_train)

print("Linear Regression Metrics:")
print(f"Mean Absolute Error (Test): {mae_test}")
print(f"Mean Absolute Error (Train): {mae_train}")
print(f"Mean Squared Error (Test): {mse_test}")
print(f"Mean Squared Error (Train): {mse_train}")
print(f"R-squared (Test): {r2_test}")
print(f"R-squared (Train): {r2_train}")
```

```python
y_pred_combined = pd.Series(index=df2.index, dtype=float)

y_pred_combined.loc[train_indices] = y_pred_train
y_pred_combined.loc[test_indices] = y_pred_test

df2['predicted_LR'] = y_pred_combined

print(df2.head())
```

```python
param_grid = {
    'kernel': ['linear', 'rbf'],
    'C': [0.1, 1, 10, 100],
    'gamma': ['scale', 'auto']
}

svm_regressor = SVR()

grid_search = GridSearchCV(estimator=svm_regressor, param_grid=param_grid, cv=25, scoring='neg_mean_absolute_error', n_jobs=-1)
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
best_score = -grid_search.best_score_

print("Best Parameters:", best_params)
print("Best Mean Absolute Error:", best_score)
```

```python
y_pred_test = grid_search.best_estimator_.predict(X_test)
y_pred_train = grid_search.best_estimator_.predict(X_train)

mae_test = mean_absolute_error(y_test, y_pred_test)
mae_train = mean_absolute_error(y_train, y_pred_train)

mse_test = mean_squared_error(y_test, y_pred_test)
mse_train = mean_squared_error(y_train, y_pred_train)

r2_test = r2_score(y_test, y_pred_test)
r2_train = r2_score(y_train, y_pred_train)

print("\nSupport Vector Machine Regressor Metrics:")
print(f"Mean Absolute Error (Test): {mae_test}")
print(f"Mean Absolute Error (Train): {mae_train}")
print(f"Mean Squared Error (Test): {mse_test}")
print(f"Mean Squared Error (Train): {mse_train}")
print(f"R-squared (Test): {r2_test}")
print(f"R-squared (Train): {r2_train}")
```
