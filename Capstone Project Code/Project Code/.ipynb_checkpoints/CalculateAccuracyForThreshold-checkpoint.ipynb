```python
import pandas as pd
df = pd.read_csv("C:/Users/abdul/Business Intelligence Project/Formatted_Data/MachineLearningResults.csv")
```

```python
df.head()
```

```python
tolerances = [0.05, 0.10, 0.15, 0.20, 0.25]
for i in tolerances:
    tolerance = i
    lower_bound = df['Actual Cost'] * (1 - tolerance)
    upper_bound = df['Actual Cost'] * (1 + tolerance)
    
    # Check if predictions are within the acceptable range
    df['DT_within_range'] = df['predicted_DT'].between(lower_bound, upper_bound)
    df['RF_within_range'] = df['predicted_RF'].between(lower_bound, upper_bound)
    df['LR_within_range'] = df['predicted_LR'].between(lower_bound, upper_bound)
    df['SVM_within_range'] = df['predicted_SVM'].between(lower_bound, upper_bound)
    
    models = ['DT', 'RF', 'LR', 'SVM']
    stats = {}
    
    for model in models:
        within_range = df[f'{model}_within_range']
        stats[model] = {
            'sum': within_range.sum(),
            'mean': within_range.mean() * 100,
            'wrong': len(df) - within_range.sum()
        }
    
    print(f"For tolerance: {tolerance*100}%")
    for model, stat in stats.items():
        print(f"{model}: Sum = {stat['sum']}, Accuracy = {stat['mean']:.2f}%, Wrong = {stat['wrong']}")
```
